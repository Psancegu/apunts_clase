### Pregunta P3 (1 punt)

Fes una funció que trobi el nombre n més gran tal que compleixi les tres condicions següents:

+ 1. n < 10.000
+ 2. n = m * m per a algun m
+ 3. n = k! + 1 per a algun k

Utilitza la plantilla següent (et donem una part feta també)

```python
import math
def numero_3_condicions():
    """
    Aquesta funció troba els numeros que compleixen les tres condicions demanades
    
    Returns
    -------
    buscat: int
    """
    l = []
    l2 = []
    f = set()
    for i in range(10):
        f.add(math.factorial(i)+1)
    for i in f:
      l.append(i)
    print(l)
    for n in l:
      if(n > 10000):
        l.remove(n)
    print(l)
    for x in range(len(l)):
      for y in range(100):
        if(l[x] == y*y):
          l2.append(l[x])
    print(l2)
    for t in range(len(l2)-1):
      if(math.factorial(t)+1 != l2):
        l2.remove(l2[t])
    print(l2)
    buscat = l2[0]

    return buscat  
```
