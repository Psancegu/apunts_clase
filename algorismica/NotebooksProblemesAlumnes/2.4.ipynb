### 2.4 Pendent Recta

Considera dos punts del pla amb coordenades $(x_1,y_1)$ i $(x_2,y_2)$. El pendent de la recta que passa per aquests dos punts es defineix com:
$$
m = \frac{y_2-y_1}{x_2-x_1}
$$
Escriu una funció que el calculi.

```python
def pendent(x1,y1,x2,y2):
    if x2 == x1:
        return float('inf')  
    
    m = (y2 - y1) / (x2 - x1)
    return m
    """
    Aquest programa calcula la pendent d'una recta donats dos punts
    
    Parameters
    ----------
    x1, y1: float, float
        Primer punt
    x2, y2: float, float
        Segon punt
    
    Returns
    -------
    m: float
        Pendent
    """
```

```python
assert pendent(1,1,4,4) == 1
assert pendent(4,5,2,9) == -2
assert pendent(1,0,4,0) == 0.
assert pendent(0,5,0,9) == float('inf')
```

Escriu una extensió que calculi l'equació de la recta que passa per dos punts donats. L'equació de la recta es defineix com $y = mx + n$, on $m$ és la pendent i $n$ l'ordenada d'origen. El programa ha de retornar una string amb l'equació de la recta.

```python
def equacio(x1,y1, x2,y2):
    if x2 == x1:
        m = float('inf')  
    else:
        m = (y2 - y1) / (x2 - x1)

    n = y1 - m*x1 
    return 'y = {} * x + {}'.format(m, n)
    """
    Aquest programa calcula l'equació de la recta que pasa per dos punts.
    
    Parameters
    ----------
    x1, y1: float, float
        Primer punt
    x2, y2: float, float
        Segon punt
    
    Returns
    -------
    equacio: string
        Equació de la recta
    """
```

```python
assert equacio(1,1,4,4) == 'y = 1.0 * x + 0.0'
assert equacio(4,5,2,9) == 'y = -2.0 * x + 13.0'
assert equacio(-1,-3,3,5) == 'y = 2.0 * x + -1.0'
```
