### 2.2 Inversió econòmica

El següent codi calcula el valor final d'una inversió considerant un interès anual fix. 

```python
def futval(inversio, interes):
    principal = inversio
    for i in range(10):
        principal = principal * (1 + interes)
    return principal
```

A partir d'aquest codi, escriu una altra funció que donada una inversió inicial, l'interès, el nombre d'anys i la comissió del banc calculi el valor final de la inversió. 

```python
def futval2(inicial, anys, interesPerCent, comissioPerCent):
    """
    Aquest programa calcula el valor d'una inversió amb interès compost anual
    i una comissió bancària aplicada cada any després d'aplicar l'interès.

    Parameters
    ----------
    inicial: float
        Inversió inicial.
    anys: int
        Nombre d'anys de la inversió.
    interesPerCent: float
        Percentatge d'interès anual (en decimal).
    comissioPerCent: float
        Percentatge de comissió aplicada cada any (en decimal).

    Returns
    -------
    diners: float
        Valor final de la inversió després del període amb interès i comissió anual.
    """
    
    interes = interesPerCent / 100  # Convertim el percentatge d'interès a decimal
    comissio = comissioPerCent / 100  # Convertim la comissió a decimal
    
    principal = inicial
    
    for i in range(anys):
        principal = principal * (1 + interes)  # Afegim l'interès anual
        principal = principal * (1 - comissio)  # Descomptem la comissió bancària anual

    return principal

futval2(200,10,0.8,0.3)
```

```python
assert abs(futval2(200,10,0.8,0.3) - 49986.54) < 0.1
assert abs(futval2(100,20,0.8,0.3) - 8923765.35) < 0.1
assert abs(futval2(100,10,0.6,0.3) -  7696.58) < 0.1
assert abs(futval2(100,10,0.8,0.5) - 17852.33) < 0.1
```
