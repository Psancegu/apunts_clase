Fes una funció que trobi el nombre $n$ més gran tal que compleixi les tres condicions seg¨üents:
+ $n<10000$
+ $n = m^2$ per algun $m$
+ $n = k! + 1$ per algun $k$

```python
import math
def numero_3_condicions():
    quadrats_perfectes = []
    for i in range(1,100):
        quadrats_perfectes.append(pow(i, 2))
    
    nombres = []
    for i in range(1,10):
        if math.factorial(i) +1 in quadrats_perfectes:
            nombres.append(math.factorial(i) +1)
        else:
            continue

    return max(nombres)


    """
    Aquesta funció troba els numeros que compleixen les tres condicions demanades
    
    Returns
    -------
    buscat: int
    """  
```

```python
assert numero_3_condicions() == 5041
```

Fes una funció que trobi tots els nombres de setmanes $n$ que tinguin un nombre de segons $s$ tal que $s=k!$ per algun $k<100$.

```python
import math

def setmanes_segons():
    factorials = []
    for i in range(1,100):
        factorials.append(math.factorial(i))
    s = 7*24*3600
    resultats = []
    for i in factorials:
        N = i / s
        if N.is_integer() and N>0:
            resultats.append(int(N))
        else:
            continue

    return resultats
    """
    Aquesta funció troba tots els nombres de setmanes N que tenen 
    un nombre de segons S tal que S = K! per algun k < 100
    
    Returns
    -------
    res: list
    """
setmanes_segons()
```

```python
assert setmanes_segons() == [6, 66, 792]
```

Donat un nombre $n$, troba dos nombres positius tals que la suma dels seus quadrats sigui $n$, o si no n'hi ha que retorni $(-1, -1)$.

```python
import math
def quadrats(n):
    for i in range(1,n-1):
        for j in range(i,n):
            if (pow(i, 2)+pow(j, 2)) == n:
                return (i,j)
            else: 
                continue
    return (-1,-1)
    """
    Aquesta funció troba dos nombres positius tals que la suma 
    dels seus quadrats sigui n.
    
    Parameters
    ----------
    n: int
    
    Returns
    -------
    quadrats: (int, int)
    """
print(quadrats(125))
```

```python
assert quadrats(125) == (2,11)
assert quadrats(106) == (5,9)
assert quadrats(81) == (-1,-1)
```

Els nombres naturals menors que 10 que són múltiples de 3 o 5 són 3, 5, 6 i 9. La seva suma és 23. Troba la suma de tots els nombres naturals menors que 1000 que són múltiples de 3 o 5.

```python
def multiples():
    sumatori = 0
    for i in range(3,1000):
        if i % 3 == 0 or i % 5 == 0:
            sumatori += i

    return sumatori
    """
    Aquesta funció retorna la suma de tots els naturals que 
    són múltiples de 3 o 5 i que son menors de 1000.
    
    Returns
    -------
    suma : int
    """
```

```python
assert multiples() == 233168
```
