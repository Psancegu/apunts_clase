### 2.8 Factorial menor que un nombre

Escriu una funció que retorni tots els valors menors que un nombre donat i que són factorials d'algun nombre natural. Fes-ho sense usar cap funció específica de Python per calcular factorials.

```python
def factorial_menor(num):
    llista_factorials = []
    factorial = 1
    for i in range(num):
        factorial = factorial*(i + 1)
        if factorial < num:
            llista_factorials.append(factorial)
        
    return llista_factorials
    """
    Aquesta funció calcula tots els factorials menors a un nombre
    
    Parameters
    ----------
    num: int
    
    Returns
    -------
    factorials: list
    """
    
```

```python
assert factorial_menor(20) == [1, 2, 6]
assert factorial_menor(50) == [1, 2, 6, 24]
```
