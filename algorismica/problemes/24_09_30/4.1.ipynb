### 4.1 Acronims

Escriu una funció que a partir d'una frase sense punts, comes, accents, números ni cap signe de puntuació, imprimeixi l'acrònim corresponent.

```python
def acronim(frase):
    acronim = ""
    llista_paraules = frase.split(" ")
    for paraula in llista_paraules:
        acronim = acronim + paraula[0].upper() # O acronim += paraula[0].upper()
    return acronim

```

```python
assert acronim("Hola que tal") == 'HQT'

assert acronim("Que estic fent?") == 'QEF'
```
