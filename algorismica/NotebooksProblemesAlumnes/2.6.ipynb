### 2.6 Palindrom

Donada una cadena de caràcters, troba tots els palíndroms generables a partir de les seves lletres. Pots fer servir la llibreria `itertools` per generar totes les possibles combinacions.

```python
import itertools

def generar_palindroms(cadena):
    llista_lletres = []
    for lletra in cadena:
        llista_lletres.append(lletra)
    
    permutacions = itertools.permutations(llista_lletres)
    palindroms = []

    for combinacio in permutacions:
        if combinacio == combinacio[::-1] and combinacio not in palindroms:
            palindroms.append(combinacio)
        else:
            continue

    for i in palindroms:
        print("".join(map(str, i)))
    """
    Aquesta funció crea tots els palíndroms possibles amb les lletres de la cadena donada.
    
    Parameters
    ----------
    cadena: string
        La cadena donada
    
    Returns
    -------
    None
    """  
generar_palindroms('aarrj')
generar_palindroms('aann')
generar_palindroms('aabtb')
```
